```python
# needed library
import pandas as pd
import numpy as np
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression , Ridge , Lasso
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_squared_error
from ipywidgets import Dropdown, FloatSlider, IntSlider, interact
```

```python
# data connection
df = pd.read_csv('/content/drive/MyDrive/DEPI/Salary_Data.csv')
df.info()
print()
print(df.head())
print()
print(df.describe())
```

```python
#data spliting
x = df.iloc[:,0].values
y = df.iloc[:,1].values
x = x.reshape(-1,1)
y = y.reshape(-1,1)
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
model = make_pipeline(
    StandardScaler(),
    LinearRegression()
    )
model.fit(x_train,y_train)
```

```python
y_pred_training = model.predict(x_test)
print("Training MAE:", mean_squared_error(y_test , y_pred_training))
```

```python
def make_prediction(YearsExperience):
    data = {
        "YearsExperience" : YearsExperience
    }
    df = pd.DataFrame(data , index=[0])
    prediction = model.predict(df)
    return f"Predicted apartment price: ${prediction}"
```

```python
interact(
    make_prediction,
    YearsExperience=FloatSlider(
        min=x_train[:, 0].min(),  # Access the first column using integer indexing
        max=x_train[:, 0].max(),  # Access the first column using integer indexing
        step=1,
        value=x_train[:, 0].mean(),
    ),
);
```
